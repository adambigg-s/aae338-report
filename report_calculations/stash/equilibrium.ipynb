```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma = 5.67e-8 
solar_constant = 1361
k = 1.5 
```

```python
T_eq = (solar_constant / (4 * sigma))**0.25

radii = np.linspace(1, 1000, 500)
temperatures = np.full_like(radii, T_eq)

plt.figure(figsize=(8,5))
plt.plot(radii, temperatures, label='Equilibrium Temp')
plt.xlabel("Radius of Dome (m)")
plt.ylabel("Equilibrium Temperature (K)")
plt.title("Equilibrium Temperature vs Dome Radius")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print(f"Equilibrium Temperature: {T_eq:.2f} K")
```

```python
T_outer = (solar_constant / (4 * sigma))**0.25

radii = np.linspace(10, 1000, 500)
R_outer = radii
R_inner = 0.9 * radii

P_absorbed = np.pi * R_outer**2 * solar_constant

T_inner = T_outer - (P_absorbed * (1/R_inner - 1/R_outer)) / (4 * np.pi * k)

plt.figure(figsize=(8,5))
plt.plot(radii, T_inner, label="Internal Air Temp (K)")
plt.axhline(T_outer, label=f"Outer Temp {T_outer:.2f} K")
plt.xlabel("Dome Radius (m)")
plt.ylabel("Internal Air Temperature (K)")
plt.title("Internal Temperature vs Dome Radius (with 10% wall)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
